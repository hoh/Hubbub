```python
import numpy as np
import pandas as pd
```

```python
from hubbub.generator.generator import Simulator
from hubbub.generator.heartbeat import HeartBeatSimulator
from hubbub.datasets.simulations import SIMPLE_LOG
```

# Running simulator

```python
result_sm = Simulator(SIMPLE_LOG).run()
results_HB = [
    HeartBeatSimulator(SIMPLE_LOG).run() for i in xrange(5)
#    HeartBeatSimulator(SIMPLE_LOG).run(delay=lambda: 5) for i in xrange(10)
    ]
```

## Converting results to timestamps for plotting:

```python
import time
def timestamp(n):
    unix_time = time.mktime(n.timetuple())
    return unix_time
```

```python
r_real = pd.DataFrame(
    [(0, timestamp(i['date']), 'SIMPLE_LOG') for i in SIMPLE_LOG],
    columns=('dummy', 'timestamp', 'source'),
    )
r_real
```

```python
r_dummyHB = [
    pd.DataFrame(
        [(1, timestamp(i['date']), 'HB{}'.format(index)) for i in r],
        columns=('dummy', 'timestamp', 'source'),
        )
    for index, r in enumerate(results_HB)
    ]
r_dummyHB
```

# Analyzing delays

## Plotting

```python
r = r_dummyHB[0]
figure()
plot(r[1:101] - r[0:100], 'o-')
show()

figure()
for r in r_dummyHB:
    plot(r[1:101] - r[0:100], 'o')
show()
```

```python
r_mixed = [
    np.concatenate((r_real, r))
    for r in r_dummyHB
]
for r in r_mixed:
    r.sort()
```

```python
r = r_mixed[0]
figure()
plot(r[1:101] - r[0:100], 'o-')
show()

figure()
for r in r_mixed:
    plot(r[1:101] - r[0:100], 'o')
show()
```
