```python
import numpy as np
import pandas as pd
```

```python
from hubbub.generator.generator import Simulator
from hubbub.generator.heartbeat import HeartBeatSimulator
from hubbub.datasets.simulations import SIMPLE_LOG
```

# Running simulator

```python
result_sm = Simulator(SIMPLE_LOG).run()
results_HB = [
    HeartBeatSimulator(SIMPLE_LOG).run() for i in xrange(5)
#    HeartBeatSimulator(SIMPLE_LOG).run(delay=lambda: 5) for i in xrange(10)
    ]
```

## Converting results to timestamps for plotting:

```python
import time
def timestamp(n):
    unix_time = time.mktime(n.timetuple())
    return unix_time
```

```python
r_real = pd.DataFrame(
    [(0, timestamp(i['date']), 'SIMPLE_LOG') for i in SIMPLE_LOG],
    columns=('dummy', 'timestamp', 'source'),
    )
r_real
```

```python
r_dummyHB = [
    pd.DataFrame(
        [(1, timestamp(i['date']), 'HB{}'.format(index)) for i in r],
        columns=('dummy', 'timestamp', 'source'),
        )
    for index, r in enumerate(results_HB)
    ]
r_dummyHB
```

# Analyzing delays

## Plotting

```python
r = r_dummyHB[0]
figure()
plot(r['timestamp'][:100].diff(), 'o-')
show()

figure()
for r in r_dummyHB:
    plot(r['timestamp'][:100].diff(), 'o')
show()
```

```python
r_mixed = [
    pd.concat((r_real, r))
    for r in r_dummyHB
]
for r in r_mixed:
    r.sort('timestamp', inplace=True)
    r['delta'] = r['timestamp'].diff()
r_mixed[0]
```

```python
# Average delay
for r in r_mixed:
    print 'all  mean', r['delta'].mean()
    print 'all  std ', r['delta'].std()
    print 'dumm mean', r[r['dummy'] == 0]['delta'].mean()
    print 'dumm std ', r[r['dummy'] == 0]['delta'].std()
    print 'real mean', r[r['dummy'] == 1]['delta'].mean()
    print 'real std ', r[r['dummy'] == 1]['delta'].std()

bar(
    range(0, len(r_mixed)*2, 2), 
    [r[r['dummy'] == 0]['delta'].mean() for r in r_mixed], 
    yerr=[r[r['dummy'] == 0]['delta'].std() for r in r_mixed], 
    color='y',
    )

bar(
    range(1, len(r_mixed)*2, 2), 
    [r[r['dummy'] == 1]['delta'].mean() for r in r_mixed], 
    yerr=[r[r['dummy'] == 1]['delta'].std() for r in r_mixed], 
    color='r',
    )
```

```python
r = r_mixed[0]
figure()
plot(r['delta'][:100], 'o-')
show()

figure()
for r in r_mixed:
    plot(r['delta'][:100], 'o')
show()
```
