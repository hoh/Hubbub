# Running simulator

```python
result_sm = Simulator(SIMPLE_LOG).run()
results_HB = [
    HeartBeatSimulator(SIMPLE_LOG).run() for i in xrange(5)
#    HeartBeatSimulator(SIMPLE_LOG).run(delay=lambda: 5) for i in xrange(10)
    ]

results_HB[0][:2]
```

## Converting results to timestamps for plotting:

```python
import time
def timestamp(n):
    unix_time = time.mktime(n.timetuple()) + n.microsecond/1000000.
    return unix_time
```

```python
r_real = pd.DataFrame(
    [(0, timestamp(i[0]), 'SIMPLE_LOG', i[1]) for i in SIMPLE_LOG],
    columns=('dummy', 'timestamp', 'source', 'length'),
    )
r_real.head()
```

```python
r_dummyHB = [
    pd.DataFrame(
        [(1, timestamp(i[0]), 'HB{}'.format(index), i[1]) for i in r],
        columns=('dummy', 'timestamp', 'source', 'length'),
        )
    for index, r in enumerate(results_HB)
    ]
r_dummyHB[0].head()
```

# Analyzing delays

## Computing deltas for n-3 to n+3

```python
r_mixed = [
    pd.concat((r_real, r))
    for r in r_dummyHB
]
for r in r_mixed:
    r.sort('timestamp', inplace=True)
    r['d-3'] = r['timestamp'].diff(periods=+3)
    r['d-2'] = r['timestamp'].diff(periods=+2)
    r['d-1'] = r['timestamp'].diff(periods=+1)
    r['d+1'] = -r['timestamp'].diff(periods=-1)
    r['d+2'] = -r['timestamp'].diff(periods=-2)
    r['d+3'] = -r['timestamp'].diff(periods=-3)

r_mixed[0].head(10)
```

## Comparing real and dummy delays

```python
# Average delay
for r in r_mixed:
    print 'all  mean', r['d-1'].mean()
    print 'all  std ', r['d-1'].std()
    print 'dumm mean', r[r['dummy'] == 1]['d-1'].mean()
    print 'dumm std ', r[r['dummy'] == 1]['d-1'].std()
    print 'real mean', r[r['dummy'] == 0]['d-1'].mean()
    print 'real std ', r[r['dummy'] == 0]['d-1'].std()
    print

bar(
    range(0, len(r_mixed)*3, 3), 
    [r['d-2'].mean() for r in r_mixed], 
    yerr=[r['d-2'].std() for r in r_mixed], 
    color='g',
    )
    
bar(
    range(1, len(r_mixed)*3, 3), 
    [r[r['dummy'] == 1]['d-2'].mean() for r in r_mixed], 
    yerr=[r[r['dummy'] == 1]['d-2'].std() for r in r_mixed], 
    color='y',
    )

bar(
    range(2, len(r_mixed)*3, 3), 
    [r[r['dummy'] == 0]['d-2'].mean() for r in r_mixed], 
    yerr=[r[r['dummy'] == 0]['d-2'].std() for r in r_mixed], 
    color='r',
    )
```
